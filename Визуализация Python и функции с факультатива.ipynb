```python
def check_two_samples(r1, r2, alpha = 0.05):
    ttest, pval = ttest_ind(r1, r2)
    
    if pval > alpha:
        print("Гипотеза H0 принимается")
    else:
        print("Гипотеза H0 не принимается")
```

```python
for i in range(1,11):
    print("Alpha = ",0.01*i)
    check_two_samples(r1s, r0s, alpha = 0.01*i)
    print('\n')
```

```python
def check_all_alpha(r1, r2, x, y):
    for i in range(x, y + 1):
        print("Alpha = ",0.01*i)
        check_two_samples(r1, r2, alpha = 0.01*i)
```

```python
a = 24
b = 32

check_all_alpha(r1s, r0s, a, b)
```

```python
def avg_time(d, t1, t2, t_gr):
    
    d[t1] = pd.to_datetime(d[t1])
    d[t2] = pd.to_datetime(d[t2])
    
    d['delta'] = (d[t2] - d[t1]).dt.components['days']
    df_grouped = d.groupby(t_gr).mean()['delta'].reset_index()

    print(df_grouped)
    max_t1 = d[t1].max()
    min_t1 = d[t1].min()
    date_span = max_t1 - min_t1
    
    cnt_obs = len(d)

    print(max_t1)
    print(min_t1)
    print(date_span)
    print(cnt_obs)
    return df_grouped
```
