```python
####################################################
##### Здесь я делаю аналог SQLовского case when
###################################################

a=40

### case when ...
if a < 5:
    print('a < 5')
#### when ...
elif a < 15:
    print('a between 5 and 15 ')
#### else 
else:
    print('a>15')
```

```python
##### Цикл с условием

for i in range(1,100):
    if i > 6:
        print(i)
    else:
        print(i**2)
```

```python
###### Множественные циклы

for i in range(0,15):
    for j in range(0,15):
        print('i = ',i)
        print('j = ',j)
```
